# Stemming & Lemmatization 

## Overview
- **Stemming:** Reduces words to their root form (may not be real words)  
- **Lemmatization:** Reduces words to their dictionary form (valid words)



```python
from nltk.stem import PorterStemmer, WordNetLemmatizer
import nltk
nltk.download('wordnet')

words = ["running", "jumps", "easily"]

# Stemming
stemmer = PorterStemmer()
print([stemmer.stem(w) for w in words])

# Lemmatization
lemmatizer = WordNetLemmatizer()
print([lemmatizer.lemmatize(w, pos='v') for w in words])
```
