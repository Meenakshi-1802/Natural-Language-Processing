# 1. Text Cleaning -

## Overview
Text cleaning is the first step in NLP. It removes unwanted characters, punctuation, numbers, and symbols from raw text to prepare it for analysis.

## Steps
1. Convert text to lowercase  
2. Remove punctuation  
3. Remove numbers  
4. Remove special characters  



```python
import re

text = "Hello World! NLP is amazing. Let's clean this text. 123"

# Convert to lowercase
text = text.lower()

# Remove punctuation and special characters
text = re.sub(r'[^a-z\s]', '', text)

print(text)
```
